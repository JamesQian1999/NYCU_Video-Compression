```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def psnr(output, target):
    mse = np.mean((output - target) ** 2)
    if(mse == 0):
        return 100
    max_pixel = 255.
    psnr = 10 * np.log10(max_pixel**2 / mse)
    return np.mean(psnr)
```

```python
img = Image.open('lena.png')
img = img.convert('L')
r = 256
img = img.resize((int(r), int(r)))
image = np.asarray(img)
```

```python
dct_1d  = np.load('dct_1d.npy')
idct_1d = np.load('idct_1d.npy')

print(f"PSNR: {psnr(idct_1d.astype(int), image):.3f} dB")

plt.figure(num='astronaut',figsize=(30,30))

plt.subplot(1,3,1)
plt.title('Original Image',fontsize = 25)
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title('DCT 1D',fontsize = 25) 
plt.imshow(dct_1d*700, cmap='gray', vmin=dct_1d.min(), vmax=dct_1d.max())
plt.axis('off')

plt.subplot(1,3,3)
plt.title('IDCT 1D',fontsize = 25) 
plt.imshow(idct_1d, cmap='gray')
plt.axis('off')

plt.subplots_adjust(wspace=0.01)
```
